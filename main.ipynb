```python
import pandas as pd

# Load the datasets
visits = pd.read_csv("visits_-_analytics_hw.csv")
subscriptions = pd.read_csv("subscriptions_-_analytics_hw.csv")

# Show the first few rows of both datasets
print("Visits Data:")
print(visits.head())  # Show first few rows of visits
print("\nSubscriptions Data:")
print(subscriptions.head())  # Show first few rows of subscriptions
```

```python
# Check for missing data
print("Missing values in visits data:")
print(visits.isnull().sum())  # Shows number of missing values in each column

print("\nMissing values in subscriptions data:")
print(subscriptions.isnull().sum())
```

```python
print(visits.columns)
print(subscriptions.columns)
```

```python
visits.rename(columns={"ACCOUNT_ID": "account_id"}, inplace=True)
subscriptions.rename(columns={"ACCOUNT_ID": "account_id"}, inplace=True)
```

```python
merged_data = pd.merge(visits, subscriptions, on="account_id", how="left")
print(merged_data.head())
```

```python
# 1.Q  Revenue by Region  1 Q
import matplotlib.pyplot as plt

revenue_by_region = merged_data.groupby('REGION')['REVENUE'].sum()
print(revenue_by_region)

# Optionally, visualize the result
revenue_by_region.plot(kind='bar', title='Revenue by Region', color='skyblue')
plt.ylabel('Revenue (USD)')
plt.show()
```

```python
# 2.Q  Revenue per Visitor by Channel
import matplotlib.pyplot as plt

# Calculate revenue per visitor
revenue_per_visitor = merged_data.groupby('CHANNEL').agg({'REVENUE': 'sum', 'VISIT_ID': 'count'})
revenue_per_visitor['REVENUE_PER_VISITOR'] = revenue_per_visitor['REVENUE'] / revenue_per_visitor['VISIT_ID']

# Sort values by revenue per visitor
sorted_revenue_per_visitor = revenue_per_visitor.sort_values(by='REVENUE_PER_VISITOR', ascending=False)

# Find the channel with the highest revenue per visitor
highest_channel = sorted_revenue_per_visitor.index[0]
highest_value = sorted_revenue_per_visitor['REVENUE_PER_VISITOR'].iloc[0]

print(sorted_revenue_per_visitor)
print(f"The channel with the highest revenue per visitor is '{highest_channel}' with a value of {highest_value:.2f}")

# Visualize the result
sorted_revenue_per_visitor['REVENUE_PER_VISITOR'].plot(kind='bar', title='Revenue per Visitor by Channel', color='orange')
plt.ylabel('Revenue per Visitor')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate conversion rate by landing page
conversion_rate = merged_data.groupby('LANDING_PAGE').apply(
    lambda x: x['account_id'].nunique() / x['VISIT_ID'].nunique() * 100
)

# Sort conversion rates in descending order
sorted_conversion_rate = conversion_rate.sort_values(ascending=False)

# Find the landing page with the highest conversion rate
highest_page = sorted_conversion_rate.index[0]
highest_rate = sorted_conversion_rate.iloc[0]

print(sorted_conversion_rate)
print(f"The landing page with the highest conversion rate is '{highest_page}' with a rate of {highest_rate:.2f}%.")

# Visualize the result
sorted_conversion_rate.plot(kind='bar', title='Conversion Rate by Landing Page', color='green')
plt.ylabel('Conversion Rate (%)')
plt.show()
```

```python
# 4. Q  Top Converting Landing Page by Region
import matplotlib.pyplot as plt

top_page_by_region = merged_data.groupby(['REGION', 'LANDING_PAGE']).apply(
    lambda x: x['account_id'].nunique() / x['VISIT_ID'].nunique() * 100
)
print(top_page_by_region.sort_values(ascending=False))

# Optionally, visualize the result for the top converting landing page per region
top_page_by_region.unstack().plot(kind='bar', stacked=False, title='Top Converting Landing Page by Region')
plt.ylabel('Conversion Rate (%)')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate direct subscription rate by region
direct_sub_rate = merged_data[merged_data['TRIAL_START_DATE'].isnull()].groupby('REGION').apply(
    lambda x: x['account_id'].nunique() / x['VISIT_ID'].nunique() * 100
)

# Sort direct subscription rates in descending order
sorted_direct_sub_rate = direct_sub_rate.sort_values(ascending=False)

# Find the region with the highest direct subscription rate
highest_region = sorted_direct_sub_rate.index[0]
highest_rate = sorted_direct_sub_rate.iloc[0]

print(sorted_direct_sub_rate)
print(f"The region with the highest direct subscription rate is '{highest_region}' with a rate of {highest_rate:.2f}%.")

# Visualize the result
sorted_direct_sub_rate.plot(kind='bar', title='Direct Subscription Rate by Region', color='blue')
plt.ylabel('Direct Subscription Rate (%)')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate trial start rate by region
trial_start_rate = merged_data.groupby('REGION').apply(
    lambda x: x['TRIAL_START_DATE'].notnull().sum() / x['VISIT_ID'].nunique() * 100
)

# Sort trial start rates in descending order
sorted_trial_start_rate = trial_start_rate.sort_values(ascending=False)

# Find the region with the highest trial start rate
highest_region = sorted_trial_start_rate.index[0]
highest_rate = sorted_trial_start_rate.iloc[0]

print(sorted_trial_start_rate)
print(f"The region with the highest trial start rate is '{highest_region}' with a rate of {highest_rate:.2f}%.")

# Visualize the result
sorted_trial_start_rate.plot(kind='bar', title='Trial Start Rate by Region', color='purple')
plt.ylabel('Trial Start Rate (%)')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate trial conversion rate by region
trial_conversion_rate = merged_data.groupby('REGION').apply(
    lambda x: x['SUBSCRIPTION_START_DATE'].notnull().sum() / x['TRIAL_START_DATE'].notnull().sum() * 100
)

# Sort trial conversion rates in descending order
sorted_trial_conversion_rate = trial_conversion_rate.sort_values(ascending=False)

# Find the region with the highest trial conversion rate
highest_region = sorted_trial_conversion_rate.index[0]
highest_rate = sorted_trial_conversion_rate.iloc[0]

print(sorted_trial_conversion_rate)
print(f"The region with the highest trial conversion rate is '{highest_region}' with a rate of {highest_rate:.2f}%.")

# Visualize the result
sorted_trial_conversion_rate.plot(kind='bar', title='Trial Conversion Rate by Region', color='red')
plt.ylabel('Trial Conversion Rate (%)')
plt.show()
```
